```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
# print(data)
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
print(data['d'])
# print(data)
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()

plt.scatter('a', 'b', marker='^', c='#7f7f7f', s='d', data=data)
plt.scatter('b', 'a', marker='o', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

print(f(t1))

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot() by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame({
    'x': np.random.randn(100), 
    'y': np.random.randn(100)})

f, axes = plt.subplots(1, 2)
sns.set()
sns.lmplot(x='x', y='y', data=df, fit_reg=True, scatter_kws={"marker": "D", "s": 100}, ax=axes[0])
plt.title('Show regression line')
sns.lmplot(x='x', y='y', data=df, fit_reg=False, scatter_kws={"marker": "D", "s": 100}, ax=axes[1])
plt.title("Without line")
plt.show()
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
g = sns.FacetGrid(data=tips, col="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()

g = sns.FacetGrid(tips, col="time", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".3", fit_reg=False, x_jitter=.1)
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
# print(x)
# print(y)
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red', label="Fukkk")
axs[1, 1].set_title('Axis [1, 1]')
axs[1, 1].legend()

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')
    ax.label_outer()
plt.show()
```
